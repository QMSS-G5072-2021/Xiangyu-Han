### d) Age and Gender
Let's try to extract age and gender. In a number of narratives you may have noticed a pattern:

- "75 YOM, FELL DOWN STAIRS, FRACTURE LEG": A 75 year old male.
- "21MOF FINGER SHUT IN WINDOW AT HOME;NAILBED INJURY: A 21 month-old female child.
- "DX:AVULSION OF SKIN-50YOF-CUT THUMB WITH SCISSORS TONIGHT-OPENING A BOX": A 50 year old female.
- "LEFT ELBOW CONTUSION,RT SHIN LAC.PT WAS ICE SKATING AND FELL.": No information about age or gender.

You are generally looking for a pattern with a number (age), something to indicate the units, e.g. YO or YR for years old, or MO for months old, and a character that identifies the gender. So, I suggest to think of a pattern that combines these three parts.

Create variables for age and gender (make sure to convert months to years when necessary). How many men and women are in the data? Provide a line graph of the number of injuries (y-axis) vs age (x-axis) disaggregated by gender.

```python
age_gender_list = [i for i in df[0] if re.search(r'\d{1,3}\s*(YR|YO|MO)\s*(F|M)\w*\b',i)]
len(age_gender_list)
```

```python
age_gender_df = pd.DataFrame(age_gender_list,columns=['text'])
age_gender_df.head(5)
```

```python
age_gender_df['age'] = [match.group(1) for match in [re.search(r'(\d{1,3})\s*(YR|YO|MO)\s*(F|M)\w*\b',i) for i in age_gender_df['text']] if match]
age_gender_df['gender'] = [match.group(3) for match in [re.search(r'(\d{1,3})\s*(YR|YO|MO)\s*(F|M)\w*\b',i) for i in age_gender_df['text']] if match]
age_gender_df['age_unit']= [match.group(2) for match in [re.search(r'(\d{1,3})\s*(YR|YO|MO)\s*(F|M)\w*\b',i) for i in age_gender_df['text']] if match]
```

```python
# transform from month to year
age_gender_df['age'] = age_gender_df['age'].astype(str).astype(int)
age_gender_df.loc[age_gender_df['age_unit'] == 'MO', 'age'] = age_gender_df['age']/12
age_gender_df = age_gender_df.drop(columns=['age_unit'])
age_gender_df.head(5)
```

```python
age_gender_df['gender'].value_counts()
```

```python
count = age_gender_df.groupby('age').count()
```

```python
age_gender_df['count'] = age_gender_df.groupby('age')['age'].transform('count')
```

```python
fig, ax = plt.subplots(figsize = (15,7))
age_gender_df[age_gender_df['gender'] == 'M']['age'].value_counts().sort_index().plot.line(label = 'Male')
age_gender_df[age_gender_df['gender'] == 'F']['age'].value_counts().sort_index().plot.line(label = 'Female')
leg = ax.legend()
```
